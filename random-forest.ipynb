```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, auc

cat = pd.read_csv('files_for_lab/categorical.csv')
num = pd.read_csv('files_for_lab/numerical.csv')
target = pd.read_csv('files_for_lab/target.csv')
target = target['TARGET_B']
```

```python
df = pd.concat([cat, num], axis=1)
df
```

```python
df = pd.get_dummies(df, columns=['STATE', 'HOMEOWNR', 'GENDER', 'RFA_2R', 'RFA_2A', 'GEOCODE2', 'DOMAIN_A', 'DOMAIN_B'])
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=42, stratify=target)
```

```python
# Apply SMOTE to upsample the training data
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
# Train the Random Forest Classifier with default parameters
classification = RandomForestClassifier(random_state=0)
classification.fit(X_train, y_train)
```

```python
# Make predictions
predictions = classification.predict(X_test)
```

```python
# Calculate ROC AUC score
roc_auc = roc_auc_score(y_test, predictions)
```

```python
# Print ROC AUC and classification report
print(f"ROC AUC Score: {roc_auc:.2f}")
print(classification_report(y_test, predictions))
```
